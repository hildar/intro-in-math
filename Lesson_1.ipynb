Напишите код на Python, реализующий расчет длины вектора, заданного его координатами.

```python
def calc_vektor(vektor):
    if isinstance(vektor, list) or isinstance(vektor, np.ndarray):
        result = 0
        for i in vektor:
            result += i ** 2
        result = result ** 0.5
        return result
    else:
        return 'pass a list in function'
```

```python
vk1 = [-20, 20, 20]
```

```python
vk2 = np.array([20, 20, 0])
```

```python
calc_vektor(vk1)
```

```python
calc_vektor(vk2)
```

#### Задание № 3

Напишите код на Python, реализующий построение графиков:
1. окружности,
2. эллипса,
3. гиперболы.

```python
from pylab import rcParams
rcParams['figure.figsize'] = 4, 4
```

1. Окружность

```python
# R - радиус, x0, y0 - центр окружности
def circle(R, x0=0, y0=0):
    x = np.linspace(-R + x0, R + x0, 1000)
    y1 = (R**2 - (x - x0)**2)**0.5 + y0
    y2 = -(R**2 - (x - x0)**2)**0.5 + y0
    plt.plot(x, y1)
    plt.plot(x, y2)
    plt.show()
```

```python
circle(3, 4, 1)
```

2. Эллипс

```python
# a - большая полуось, b - меньшая полуось, x0, y0 - центр окружности
def ellipse(a, b, x0=0, y0=0):
    x = np.linspace(-a + x0, a + x0, 1000)
    y1 = ((1 - (x - x0)**2 / a**2) * b**2)**0.5 + y0
    y2 = -((1 - (x - x0)**2 / a**2) * b**2)**0.5 + y0
    plt.plot(x, y1)
    plt.plot(x, y2)
    plt.show()
```

```python
# для удобства просмотра эллипса поменяем размер графика
rcParams['figure.figsize'] = 6, 4
```

```python
ellipse(5, 2, 4, -2)
```

3. Гипербола

```python
# a - большая полуось, b - меньшая полуось, x0, y0 - центр окружности
def hyperbola(a, b, x0=0, y0=0):
    y = np.linspace(-b * 3 + x0, b * 3 + x0, 1000)
    x1 = (((y - y0)**2 / b**2 + 1) * a**2)**0.5 + x0
    x2 = -(((y - y0)**2 / b**2 + 1) * a**2)**0.5 + x0
    plt.plot(x1, y)
    plt.plot(x2, y)
    plt.show()
```

```python
# для удобства просмотра эллипса поменяем размер графика
rcParams['figure.figsize'] = 3, 4
```

```python
hyperbola(4, 2, 7, 7)
```
