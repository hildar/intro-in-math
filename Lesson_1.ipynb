#### Задание № 3

Напишите код на Python, реализующий построение графиков:
1. окружности,
2. эллипса,
3. гиперболы.

````python
from pylab import rcParams
rcParams['figure.figsize'] = 4, 4
````

1. Окружность

````python
# R - радиус, x0, y0 - центр окружности
def circle(R, x0=0, y0=0):
    x = np.linspace(-R + x0, R + x0, 1000)
    y1 = (R**2 - (x - x0)**2)**0.5 + y0
    y2 = -(R**2 - (x - x0)**2)**0.5 + y0
    plt.plot(x, y1)
    plt.plot(x, y2)
    plt.show()
````

````python
circle(3, 4, 1)
````

2. Эллипс

````python
# a - большая полуось, b - меньшая полуось, x0, y0 - центр окружности
def ellipse(a, b, x0=0, y0=0):
    x = np.linspace(-a + x0, a + x0, 1000)
    y1 = ((1 - (x - x0)**2 / a**2) * b**2)**0.5 + y0
    y2 = -((1 - (x - x0)**2 / a**2) * b**2)**0.5 + y0
    plt.plot(x, y1)
    plt.plot(x, y2)
    plt.show()
````

````python
# для удобства просмотра эллипса поменяем размер графика
rcParams['figure.figsize'] = 6, 4
````

````python
ellipse(5, 2, 4, -2)
````

3. Гипербола

````python
# a - большая полуось, b - меньшая полуось, x0, y0 - центр окружности
def hyperbola(a, b, x0=0, y0=0):
    y = np.linspace(-b * 3 + x0, b * 3 + x0, 1000)
    x1 = (((y - y0)**2 / b**2 + 1) * a**2)**0.5 + x0
    x2 = -(((y - y0)**2 / b**2 + 1) * a**2)**0.5 + x0
    plt.plot(x1, y)
    plt.plot(x2, y)
    plt.show()
````

````python
# для удобства просмотра эллипса поменяем размер графика
rcParams['figure.figsize'] = 3, 4
````

````python
hyperbola(4, 2, 7, 7)
````

#### Задание № 4
1. Решите систему уравнений:
``` 
y = x**2 – 1 
exp(x) + x∙(1 – y) = 1
```



````python
from scipy.optimize import fsolve
````

````python
# Нарисуем графики
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 8)
plt.grid(True)
plt.show()
````

````python
# Найдем корни
def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x * y - 1)

x1, y1 = fsolve(equations, (-2, 1))
print(x1, y1)

x2, y2 = fsolve(equations, (2, 5))
print(x2, y2)
````

2. Решите систему уравнений и неравенств:
```
y = x**2 – 1
exp(x) + x∙(1 – y) > 1
```

Выразим __y__ через оба уравнения:
```
y = x**2 - 1
y < (exp(x) + x - 1) / x
```

````python
x = np.linspace(-3, 6, 201)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 25)
plt.grid(True)
plt.show()
````

По графику видно, что корни уравнения лежат примерно в следующих пределах оси __x__:  [-2, 2] и [4, inf]

Для решения задачи найдем точки пересечения графиков и укажем более точные значения __y__:

````python
def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x * y - 1)

x1, y1 = fsolve(equations, (-2, 2))
xy = np.array([x1, y1])

x2, y2 = fsolve(equations, (2, 5))
xy = np.append([xy], [[x2, y2]], axis=0)

x3, y3 = fsolve(equations, (4, 16))
xy = np.append(xy, [[x3, y3]], axis=0)
xy
````

Таким образом получаем, что корни системы уравнений и неравенств лежат на параболе ```y = (x**2 - 1)``` в пределах следующих значений оси __x__:

````python
print(f'x: ({xy[0,0]}, {xy[1,0]})')
print(f'x: ({xy[2,0]}, inf]')
````
